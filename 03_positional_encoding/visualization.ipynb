```python
import torch
import matplotlib
import matplotlib.pyplot as plt

matplotlib.use("module://matplotlib_inline.backend_inline")

from positional_encoding import PositionalEncoding

# Step 1: Initialize positional encoding
d_model = 512
pe = PositionalEncoding(d_model, max_len=100)

# Step 2: Get tensor
pe_tensor = pe.pe.squeeze(0).detach().cpu().numpy()

# Step 3: Create figure
plt.figure(figsize=(12, 6))

# Step 4: Plot selected dimensions (better visibility)
dims = [0, 2, 4, 8,]

for i in dims:
    plt.plot(pe_tensor[:, i], label=f"Dim {i} (sin)")
    plt.plot(pe_tensor[:, i + 1], linestyle="--", label=f"Dim {i+1} (cos)")

# Step 5: Styling (important for visibility)
plt.xlabel("Position")
plt.ylabel("Value")
plt.title("Sinusoidal Positional Encoding")
plt.grid(True)
plt.legend()

# Step 6: Force render
plt.show()
plt.draw()
```
